```python
dados = dados.fillna(0)
```

```python
remov_alunos = dados.query('Nome == "Alice" | Nome == "Carlos"').index
remov_alunos
```

```python
dados.drop(remov_alunos, axis=0, inplace=True)
```

```python
selecao = dados['Aprovado'] == True
filtro_aprovados = dados[selecao]
```

```python
filtro_aprovados
```

```python
filtro_aprovados.to_csv("Alunos_aprov.csv", index=False)
```

```python
sub_nota = filtro_aprovados.query('Notas == 7.0')
sub_nota
```

```python
filtro_aprovados = filtro_aprovados.replace(7, 8)
```

```python
filtro_aprovados.to_csv("Alunos_aprov.csv", index=False)
```

```python
dados['Pontos_extras'] = dados['Notas'] * 0.4
dados.head()
```

```python
dados['Notas_finais'] = dados['Notas'] + dados['Pontos_extras']
dados.head()
```

```python
dados['Aprovados_finais'] = dados['Notas_finais'].apply(lambda x : 'True' if x > 6 else 'False' )
dados.head()
```

```python
dados.to_csv('Alunos_aprov_finais.csv', index=False)
```

```python
filtro_aprovados_finais = dados.query('Aprovados_finais == "True"')
```
