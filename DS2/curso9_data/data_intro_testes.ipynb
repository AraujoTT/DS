```python
import pandas as pd

tmdb = pd.read_csv("data/tmdb_5000_movies.csv")
tmdb.head()
```

```python
tmdb.describe()
```

```python
import seaborn as sns
```

```python
ax = sns.distplot(tmdb['vote_average'])
ax.set(xlabel = 'Nota media', ylabel = "Densidade")
ax.set_title('Média de votos em filmes no TMDB')
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize= (12,8))
ax = fig.add_axes([0, 0, 1, 1])

ax = sns.distplot(tmdb['vote_average'], norm_hist=False, kde=False)
ax.set(xlabel = 'Nota media', ylabel = "Densidade")
ax.set_title('Média de votos em filmes no TMDB', fontsize=18)

ax
```

```python
fig = plt.figure(figsize= (12,8))
ax = fig.add_axes([0, 0, 1, 1])

ax = sns.boxplot(tmdb.vote_average, orient='h')

ax.set_title('Distribuiçao de nota media de filmes no TMDB', fontsize=18)
ax.set_xlabel('Nota media', fontsize= 16)
```

Tem algo de estranho com os meus dados nao faz sentido filmes cuja a nota media é 0 ou 10 e analisando persebemos que é pela baixa quantidade de votas levando em conta a probabilidade melhor retirar esses dados para nao interferir

```python
tmdb.query("vote_average == 0").head()
```

```python
tmdb.query('vote_average == 10').head()
```

```python
tmdb_com_mais_10_votos = tmdb.query("vote_count >= 10")
tmdb_com_mais_10_votos.describe()
```

```python
fig = plt.figure(figsize= (12,8))
ax = fig.add_axes([0, 0, 1, 1])

ax = sns.distplot(tmdb_com_mais_10_votos['vote_average'], norm_hist=False, kde=False)
ax.set(xlabel = 'Nota media', ylabel = "Frequencia")
ax.set_title('Média de votos em filmes no TMDB dentre os filmes com mais de 10 votos', fontsize=18)

ax
```

```python
fig = plt.figure(figsize= (12,8))
ax = fig.add_axes([0, 0, 1, 1])

ax = sns.boxplot(tmdb_com_mais_10_votos.vote_average, orient='h')

ax.set_title('Distribuiçao de nota media de filmes no TMDB dentre os filmes com 10 ou mais votos', fontsize=18)
ax.set_xlabel('Nota media', fontsize= 16)
```

# analisaremos tambem o movielens

```python
notas  = pd.read_csv('data/ratings.csv')
notas.head()
```

```python
nota_media_por_filme = notas.groupby("movieId").mean()['rating']
```

```python
ax = sns.distplot(nota_media_por_filme.values)
ax.set(xlabel="Nota media", ylabel="Densidade")
ax.set_title("Media de votos em filmes no movielens")
```

```python
quantidade_de_votos_por_filme = notas.groupby('movieId').count()
filmes_com_pelo_menos_10_votos = quantidade_de_votos_por_filme.query('rating >=10').index
```

```python
nota_media_por_filme_com_pelo_menos_10_nomes = nota_media_por_filme.loc[filmes_com_pelo_menos_10_votos.values]
```

```python
nota_media_por_filme_com_pelo_menos_10_nomes.head()
```

```python
ax = sns.distplot(nota_media_por_filme_com_pelo_menos_10_nomes.values)
ax.set(xlabel="Nota media", ylabel="Densidade")
ax.set_title("Media de votos em filmes no movielens com 10 ou mais votos")
```

```python
fig = plt.figure(figsize= (12,8))
ax = fig.add_axes([0, 0, 1, 1])

ax = sns.boxplot(nota_media_por_filme_com_pelo_menos_10_nomes.values, orient='h')

ax.set_title('Distribuiçao de nota media de filmes no movielens dentre os filmes com 10 ou mais votos', fontsize=18)
ax.set_xlabel('Nota media', fontsize= 16)
```
