```python
import datetime
```

```python

fig = plt.figure(figsize=(15, 8))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df['data'], df['temperatura'], color = 'darkblue', lw= 0.8)

eixo.set_xlim(datetime.datetime(2014,1,1), datetime.datetime(2015,1,1))
eixo.set_title('Temperatura no momento', fontsize=20)
eixo.set_ylabel('Temperatura', fontsize=16)
eixo.set_xlabel('Data', fontsize=16)
eixo.legend(['temperatura'], loc= 'lower right', fontsize = 14 )
```

```python

fig = plt.figure(figsize=(15, 8))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df['data'], df['temperatura'], color = 'darkblue', lw= 0.8, marker = 'o')

eixo.set_xlim(datetime.datetime(2014,1,1), datetime.datetime(2015,1,1))
eixo.set_title('Temperatura no momento', fontsize=20)
eixo.set_ylabel('Temperatura', fontsize=16)
eixo.set_xlabel('Data', fontsize=16)
eixo.legend(['temperatura'], loc= 'lower right', fontsize = 14 )
```

```python

fig = plt.figure(figsize=(15, 8))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df['data'], df['temperatura'], color = 'darkblue', lw= 0.8)

eixo.set_xlim(datetime.datetime(2014,1,1), datetime.datetime(2015,1,1))
eixo.set_title('Temperatura no momento', fontsize=20)
eixo.set_ylabel('Temperatura', fontsize=16)
eixo.set_xlabel('Data', fontsize=16)
eixo.legend(['temperatura'], loc= 'lower right', fontsize = 14 )
eixo.grid(True)
```
